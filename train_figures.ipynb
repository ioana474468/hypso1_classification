```python
import pickle
import matplotlib.pyplot as plt

# Încărcarea istoricului din fișierul pickle
with open('training_history_cnn.pkl', 'rb') as f:
    history_cnn = pickle.load(f)
    
with open('training_history_enc.pkl', 'rb') as f:
    history_enc = pickle.load(f)
    
with open('training_history_transf.pkl', 'rb') as f:
    history_transf = pickle.load(f)

print("Metricile găsite în fișier:", history_cnn.keys())
print("Metricile găsite în fișier:", history_enc.keys())
print("Metricile găsite în fișier:", history_transf.keys())
```

```python
h1=history_cnn
h2=history_enc
h3=history_transf

# --- GRAFICUL 1: LOSS COMPARATIV ---
plt.figure(figsize=(10, 3))

# Adăugăm linia pentru fiecare model
plt.plot(h1['loss'], label='Hybrid 3d-2d CNN', color='blue', linewidth=2)
plt.plot(h2['loss'], label='Autoencoder classifier', color='red', linewidth=2)
plt.plot(h3['loss'], label='Transformer', color='green', linewidth=2)

plt.title('Comparație Loss între cele 3 Modele')
plt.xlabel('Epoci')
plt.ylabel('Loss')
plt.legend() 
plt.grid(True, linestyle='--', alpha=0.6)

# Salvare si afișare
plt.savefig('comparatie_loss.png', dpi=300, bbox_inches='tight')
plt.show() 
plt.close()
```

```python
print(f"Model 1 a fost antrenat pentru: {len(h1['loss'])} epoci.")
print(f"Model 2 a fost antrenat pentru: {len(h2['loss'])} epoci.")
print(f"Model 3 a fost antrenat pentru: {len(h3['loss'])} epoci.")
```

```python
# --- GRAFICUL 2: ACURATEȚE COMPARATIVĂ ---

plt.figure(figsize=(10, 3))

# Adăugăm liniile de acuratețe
plt.plot(h1['accuracy'], label='Hybrid 3d-2d CNN', color='blue', linewidth=2)
plt.plot(h2['accuracy'], label='Autoencoder classifier', color='red', linewidth=2)
plt.plot(h3['accuracy'], label='Transformer', color='green', linewidth=2)

plt.title('Comparație Acuratețe între cele 3 Modele')
plt.xlabel('Epoci')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

# Salvare si afișare
plt.savefig('comparatie_acuratete.png', dpi=300, bbox_inches='tight')
plt.show()
plt.close()
```
