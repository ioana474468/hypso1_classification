model autoencoder generic

```python

model = keras.Sequential([
    keras.layers.Input(shape=(784,)),
    # Encoder
    keras.layers.Dense(8, activation='relu'),
    keras.layers.Dense(4, activation='relu'),
    keras.layers.Dense(2, activation='relu'),
    # Bottleneck
    keras.layers.Dense(1, activation='relu'),
    # Decoder
    keras.layers.Dense(2, activation='relu'),
    keras.layers.Dense(4, activation='relu'),
    keras.layers.Dense(8, activation='sigmoid')
])

options=GraphOptions(
    layer_spacing=50
)

image = visualkeras.show(model, mode='graph', to_file='grafic_autoencoder_generic.png', options=options)
image.show()
```

grafic functie ReLU

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Definim intervalul de valori pentru axa X (de la -5 la 5)
x = np.linspace(-2, 2, 1000)

# 2. Definim funcția ReLU: f(x) = max(0, x)
relu = np.maximum(0, x)

# 3. Definim subgradientul funcției ReLU
# f'(x) = 1 dacă x > 0, altfel 0
relu_grad = np.where(x > 0, 1, 0)

# 4. Configurăm și generăm graficul
plt.figure(figsize=(10, 6))

# Trasăm funcția ReLU (linie albastră continuă)
plt.plot(x, relu, label="ReLU", color="blue", linewidth=2.5)

# Trasăm subgradientul ReLU (linie roșie întreruptă)
plt.plot(x, relu_grad, label="Subgradient ReLU", color="red", linestyle="--", linewidth=2.5)

# Personalizăm aspectul graficului
plt.title("ReLU", fontsize=24)
plt.xlabel("x", fontsize=20)
plt.ylabel("Valoare", fontsize=20)

# Adăugăm axele X și Y principale (liniile de la 0)
plt.axhline(0, color='black', linewidth=1.2)
plt.axvline(0, color='black', linewidth=1.2)

# Adăugăm un grid pentru lizibilitate
plt.grid(True, linestyle=':', alpha=0.7)

# Afișăm legenda
plt.legend(fontsize=20)

# Afișăm graficul pe ecran
plt.show()
```
